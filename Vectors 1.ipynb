```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def plotVect(st, ep) :
    plt.plot([st[0], ep[0]], [st[1], ep[1]])
   
    
```

```python
st = np.array([0, 0])
ep = np.array([1, 2])

plotVect(st, ep)

st2 = np.array([1, 2])
ep2 = np.array([5, -4])
plotVect(st2, ep2)

plotVect(st, ep2)

plt.show()
```

```python
def addAndPlot(v1, v2) :
    origin = np.array([0, 0])
    plotVect(origin, v1)
    plotVect(v1, v1+v2)
    plotVect(origin, v1+v2)
    
    plt.show()
```

```python
addAndPlot(np.array([1, 2]), np.array([4, -6]))
```

```python
def myNorm(v) :
    return math.sqrt(sum(x*x for x in v))
    
a = np.array([1, 2, 3, 4, 8])    
print (myNorm(a))
print (np.linalg.norm(a))
    
```

```python
def unitVector(x) :
    return x/np.linalg.norm(x)

print (unitVector(a))
print (np.linalg.norm(unitVector(a)))
```

```python
## Projection
o = np.array([0, 0])
r = np.array([30,10])
t = np.array([10, 15])

plotVect(o, t)
plotVect(o, r)

beta = np.dot(r, t)/np.dot(r, r)
projPt = beta * r
print (projPt)

dp = np.dot(projPt, t-projPt)

perpComp = t - projPt
print ("PERP is ", perpComp)
plotVect(o, perpComp)

plotVect(o, projPt)
plotVect(projPt, t)
#plt.show()
```
