```python
A = np.random.randint(-3, 4, (3, 3))
sym = A.T@A

## eigenvectors of symmetric matrix are pairwise orthogonal
L, V = np.linalg.eig(sym)

print (np.dot(V[:, 0], V[:, 1]))
print (np.dot(V[:, 0], V[:, 2]))
print (np.dot(V[:, 1], V[:, 2]))

print ((V.T@V).round())
```

```python
from PIL import Image

vendu = Image.open("Vendu.png")
nks = Image.open("Nks.png").resize(vendu.size)

vdata = np.array(vendu)
ndata = np.array(nks)


comb = np.hstack((vdata, ndata))
for i in range(6) :
    vd = (ndata*(i/5)) +  (vdata * ((5-i)/5))
    nd = (vdata*(i/5)) +  (ndata * ((5-i)/5))
    newd = np.hstack((vd, nd))
    comb = np.hstack((comb, newd)).astype('uint8')


#plt.imshow(vd)
plt.imshow(comb)
combIm = Image.fromarray(comb)
combIm.save("modify.png")

#print (lv[0:6])
#print (venduPixels[0:6, :, :])
#nksPixels = list(nks.getdata())

#plt.imshow(vendu)
#plt.imshow(vIm)
```
