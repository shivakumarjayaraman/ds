```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.array([1, 2, 3])
b = np.array([5, 6, 7])



def pearsonCoeff(a, b) :
    anorm = a-np.mean(a)
    bnorm = b-np.mean(b)
    
    return np.dot(anorm, bnorm)/(np.linalg.norm(anorm) * np.linalg.norm(bnorm))

print (pearsonCoeff(a, b))
```

```python
def cosineSim(a, b):
    return np.dot(a,b)/(np.linalg.norm(a) * np.linalg.norm(b))
    
print (cosineSim(a, b))
```

```python
a = np.array([0, 1, 2, 3])
b = np.array([100, 101, 102, 103])

print (pearsonCoeff(a, b), cosineSim(a, b))
print (np.corrcoef(a, b))
```

```python
## Plotting correlations (pearson and cosine)

a = np.array([0, 1, 2, 3])
xvals = range(-50, 51)

pcoeff = np.zeros(len(xvals))
cosvals = np.zeros(len(xvals))
for offset in xvals :
    b = a + offset
    pcoeff[offset+50] = pearsonCoeff(a, b)
    cosvals[offset+50] =  cosineSim(a, b)
    
plt.plot(xvals, pcoeff, '*')
plt.plot(xvals, cosvals, '-')

plt.show()
```

```python
## Edge detection

t = np.zeros(30)
t[10:21] = 1

kernel = np.array([-1, 1])

edgeVal = np.zeros(len(t))
for i in range(len(t) - 1) :
    edgeVal[i] = np.dot(t[i:i+2], kernel)
   
plt.plot(np.arange(0, 30), t, '*')
plt.plot(np.arange(0, 30), edgeVal, '-')
plt.show()
```

```python
## Low Pass Filtering using a Gaussian kernel

t = np.random.normal(loc=10, size=100)
kernel = np.array([0, .1, .3, .8, 1, .8, .3, .1, 0])
kernel = kernel/sum(kernel)
print (sum(kernel))


edgeval = np.zeros(len(t) - len(kernel) + 1)
for i in range(len(edgeval)) :
    edgeval[i] = np.dot(kernel, t[i:i+len(kernel)])
    
plt.plot(np.arange(100), t)
#plt.plot(np.arange(len(edgeval)), edgeval, '*')
```
