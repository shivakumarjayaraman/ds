# K Means Clustering


```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
#Step 1 : Initialize data points

#rng = np.random.default_rng(seed=1993)
rng = np.random.default_rng()
data = rng.random((150, 2)) * 200

plt.plot(data[:,0], data[:,1], 'bo')
```

```python
## Initialize the k centroids

k = 3

ridx = rng.choice(range(len(data)), k, replace=False)
centroids = data[ridx,:]

plt.plot(data[:,0], data[:,1], 'b*')
plt.plot(centroids[:,0], centroids[:,1], 'kD')
```

```python
## Compute the distances of each point in the dataset from each centroid
for i in range(50) :
    dists = np.zeros((data.shape[0], k))
    for ci in range(k) :
        dists[:, ci] = np.sum((data - centroids[ci,:])**2, axis=1)
    
    ## Assign each data point to the group with min distance
    groupidx = np.argmin(dists, axis=1)

    ## Compute new centroids for the k clusters
    for ki in range(k):
        centroids[ki,:] = [np.mean(data[groupidx==ki, 0]), np.mean(data[groupidx==ki, 1])]
    
```

```python
   
for ki in range(k) :
    plt.plot(data[groupidx==ki, 0], data[groupidx==ki, 1],  '*')
    

## Plot the centroids
##for ki in range(k):
##   plt.plot(centroids[ki,0], centroids[ki, 1], "kH")

plt.plot(centroids[:,0], centroids[:,1], "kD")
```
