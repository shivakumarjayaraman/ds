```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ycoords = np.linspace(-1, 1, 50)
xcoords = np.zeros(50)
plt.plot(xcoords, ycoords, '*')

for slant in range(-20, 20) :
    if slant == 0: continue
    angle = np.pi/slant
    rotMatrix = np.array([[np.cos(angle), np.sin(angle)], [-1*np.sin(angle), np.cos(angle)]])
    points = np.array(list(zip(xcoords, ycoords))).T

    rotated = rotMatrix@points
    plt.plot(rotated[0,:], rotated[1,:], '.')
```

```python
## points in a circle

angles = np.linspace(0, np.pi*2, 50)
xcoords = np.zeros(len(angles))
ycoords = np.zeros(len(angles))
for i in range(len(angles)) :
    xcoords[i] = np.cos(angles[i])
    ycoords[i] = np.sin(angles[i])
    
plt.plot(xcoords, ycoords)

points = np.array(list(zip(xcoords, ycoords))).T
rotMat = np.array([[1, 0.5], [0, 0.5]])
rotated = rotMat@points
plt.plot(rotated[0,:], rotated[1,:])
```
